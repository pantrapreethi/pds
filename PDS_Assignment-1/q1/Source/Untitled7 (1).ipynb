```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.ticker as ticker
import seaborn as sns
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df= pd.read_csv('/content/drive/MyDrive/PDS_Assignment-1/q1/Raw_Data/raw_data.csv')
df.head(10)
```

```python
df['Frailty'] = df['Frailty'].replace({'Y': 1, 'N': 0})

df.head()
```

```python
df.to_csv('/content/drive/MyDrive/PDS_Assignment-1/q1/Data_Clean/clean_data.csv')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].scatter(df['Frailty'], df['Weight'], color='blue', alpha=0.6)
axes[0].set_title('Weight vs Frailty')
axes[0].set_xlabel('Frailty')
axes[0].set_ylabel('Weight')

axes[1].scatter(df['Frailty'], df['Height'], color='green', alpha=0.6)
axes[1].set_title('Height vs Frailty')
axes[1].set_xlabel('Frailty')
axes[1].set_ylabel('Height')

axes[2].scatter(df['Frailty'], df['Grip strength'], color='red', alpha=0.6)
axes[2].set_title('Grip Strength vs Frailty')
axes[2].set_xlabel('Frailty')
axes[2].set_ylabel('Grip Strength')

plt.tight_layout()
plt.show()
save_path = os.path.join('/content/drive/MyDrive/PDS_Assignment-1/q1/Result', "whg_vs_frailty.png")
fig.savefig(save_path)
```

```python
# Set the plot style
sns.set(style="whitegrid")

# Create the box plot
plt.figure(figsize=(8, 6))
s = sns.boxplot(x='Frailty', y='Grip strength', data=df, palette='Blues_d')

# Add title and labels
plt.title('Grip Strength Distribution by Frailty')
plt.xlabel('Frailty')
plt.ylabel('Grip Strength (kg)')

# Show the plot
plt.show()
save_path = os.path.join('/content/drive/MyDrive/PDS_Assignment-1/q1/Result', "grip_vs_frailty.png")
fig.savefig(save_path)
```
